Array creation with random values

```python
import numpy as np
```

```python
#random values 

arr = np.random.rand(2,3)

print(arr)
print(arr.dtype)
```

Random integer value generator

```python
arr = np.random.randint(1,10 ,(10,10))      #random int values(star,end,shape)
print(arr)
print(arr.shape)
print(arr.dtype)
print(arr.ndim)
```

Floating type values creation

```python
#Floating type values creation
arr = np.random.uniform(30,40,(5,5))
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
```

Array creation with range

```python
arr = np.arange(1,10,1) .reshape(3 ,3)  #start end and steps 

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
```

Linespace in Numpy

```python
# np.linespace (start,end and koyta point pick korbo)
arr = np.linspace(1,10,5)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
```

Logspace in Numpy

```python
#np.logspace  (start,End(included),koyta values nAttributeError)

arr = np.logspace(0,4,5)
print(arr)
```
