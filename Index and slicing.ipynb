```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])

arr1 = np.array([[1,2,3],
                [4,5,6]])

print(arr)
print(arr.ndim)  #dimension
#indexing
print(arr[2])
#change the value of index 
arr[2] = 50
print(arr)



print(arr1)
print(arr1.ndim)

#indexing
print(arr1[0][2])
#change the value of index 
arr1[1][0] = 7
print(arr1)

```

##slicing

```python
#slicing (start,end and step) 
print(arr)
#1d array slicing
arr_mod = arr[0:3]
print(arr_mod)
```

```python
#2d array slicing

#print(arr1)

row_0 = arr1[0:1 ,]
print(row_0)

row_1 = arr1[1: ,]
print(row_1)

row_2 = arr1[1: ,1:3]
print(row_2)
```

```python
arr3 = np.arange(1,10).reshape(3,3)
print(arr3)

#slicing 

row_4 = arr3[1:, 1:3]
print(row_4)
```

Advance indexing and slicing

```python
adv_arr = np.array([1,2,3,4,5])
print(adv_arr)
adv_arr[2] = 20
print(adv_arr[0:4])
```

```python
lst = np.array([10,20,30,40,50])

values = lst[[3,4]]

print(values)
```

```python
arr_1 = np.arange(1,10,1).reshape(3,3) 
print(arr_1)
print(arr_1[[1,2],[0,1]])
#advance
print(arr_1[arr_1 >2])
#2 theke jara boro tader man 0 kore dive
arr_1[arr_1 > 2] = 0
print(arr_1)
```
