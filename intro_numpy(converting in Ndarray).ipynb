ND array

```python
import numpy as np 
arr =np.array([1,2,3,4,5]) #one demansional array 
print(type(arr))
print(arr.ndim)

#2d array 

arr1 = np.array([[1,2,3],
                [4,5,6]])

print(arr1.ndim)

#3d array 

```

Array attribute

```python
#dimension 
print(arr.ndim)
print(arr1.ndim)

#shape
print(arr.shape)
print(arr1.shape)

# size 
print(arr.size)
print(arr1.size)
```

ndarray data type

```python
arr = np.array ([1,2,3,4,5])

print(arr.dtype)
```

```python
arr1 = np.array([1,2,3,4.5]) # upcasting in array 

print(arr1.dtype)

arr1 = np.array([1,2,3,4.5,'5']) 
print(arr1.dtype)
```

Seletcing a data type for a array

arr = np.array([1,2,3],dtype =np.int8)  #Here a datatype converted into a int8 datatypes 
print(arr.dtype)

```python
arr = np.array([1,2,3])
print(arr.dtype)

arr = arr.astype(np.int32)
print(arr)                        #converting data into 64 bytes to 32 bytes
print(arr.dtype)

```

Nd array creating from exiting data

```python
#list 
import numpy as np

lst = [10,20,40,50.5]
arr = np.array(lst, dtype = np.int32)
print(arr)
print(type(arr))
print(arr.dtype)


```

```python
mixed_list =[10,10.5,True,'Hello']

arr = np.array(mixed_list)

print(arr)
print(type(arr))
print(arr.dtype)
```

```python
matrix =[[1,2,3],
        [4,5,6]]

arr = np.array(matrix)

print(arr)
print(arr.ndim)
print(arr.dtype)
print(arr.shape)
```

```python
#tuple
tple =(1,2,3)

arr = np.array(tple)
print(arr)
print(type(arr))
print(arr.ndim)
print(arr.dtype)
print(arr.shape)
```

```python
#set 
st = {1,2,3}
arr = np.array(list(st))
arr = arr.astype(np.int32)
print(arr)
print(type(arr))
print(arr.ndim)
print(arr.dtype)
print(arr.shape)
```

```python
#Dictionary 
dc ={'a': 10,'b': 20,'c': 30}
keys = dc.keys()
values = dc.values()


#keys to ndarray

arr = np.array(list(keys))
print(arr)
print(type(arr))
#values to ndarry
arr = np.array(list(values))
print(arr)
arr = arr.astype(np.int16)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.size)
```
